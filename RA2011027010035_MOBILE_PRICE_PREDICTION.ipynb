```python
#Null values removed
df1=df.dropna()
```

```python
df1.isnull().sum()
```

```python
#Apply label encoder for all the string attributes
list=["Phone Name","Rating ?/5" ,"Number of Ratings","RAM","ROM/Storage","Back/Rare Camera","Front Camera" ,"Battery" ,"Processor" ,"Price in INR","Date of Scraping"]
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in list:
    df1[i]=le.fit_transform(df1[i])
```

```python
df1
```

```python
for i in list:
  df1[i]=df1[i].astype(int)
```

```python
df1
```

```python
X=df1.drop(["Price in INR"],axis=1)
X
```

```python
y=df1.iloc[:,-2]
y
```

```python
#Apply MinMax Scaler

from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
scx=mms.fit_transform(X)
```

```python
# spliting dataset into training and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(scx,y,random_state=1,test_size=.2)
```

```python
#Construct Linear Regression Model

from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
y_pred=mlr.predict(X_test)
y_pred
```

```python
y_test
```

```python
# Find R2 value
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('r2 score for perfect model is', r2)
```

```python
#Apply 6 Fold Cross Validation
from sklearn.model_selection import cross_val_score
X=df1.drop('Price in INR',axis=1).values
y=df1[['Price in INR']]
lmr = LinearRegression()
scores = cross_val_score(lmr, X, y, scoring='r2', cv=6)
scores
```
